<h1>Cartilha lendo arquivos excel com padas</h1>

```python
import pandas as pd
```

<h1>Leitura "padrão":</h1>

```python
tabela = pd.read_excel("arquivo.xlsx")
display(tabela)
```

<h1>Lendo uma planilha("aba") específica:</h1>

```python
tabela = pd.read_excel("arquivo.xlsx", sheet_name="Planilha2")
display(tabela)
```

<h1>Lendo uma planilha("aba") específica, com o índice da aba:</h1>

```python
tabela = pd.read_excel("arquivo.xlsx", sheet_name=1)
display(tabela)
```

<h1>Selecionando as colunas, pelo índice:</h1>

```python
tabela = pd.read_excel("arquivo.xlsx", sheet_name="Planilha3", usecols=[5,6])
display(tabela)
```

<h1>Selecionando as colunas, pelo "nome da coluna":</h1>

```python
tabela = pd.read_excel("arquivo.xlsx", sheet_name="Planilha3", usecols="A,F:G")
display(tabela)
```

<h1>Buscando todas as informações de um arquivo:</h1>

```python
tabela = pd.read_excel("arquivo.xlsx", sheet_name=None)
display(tabela)
```

<h1>Buscando informações de uma tabela fora da célula "A1":</h1>

```python
tabela = pd.read_excel("arquivo.xlsx", sheet_name="Planilha4", usecols=[6, 7], skiprows=7)
display(tabela)
```

```python
tabela = pd.read_excel("arquivo.xlsx", sheet_name="Planilha5", nrows=11)
display(tabela)
```

<h1>Buscando apenas as informações desejadas:</h1>

```python
tabela = pd.read_excel("arquivo.xlsx", sheet_name=None)
display(tabela['Planilha5'])
```
